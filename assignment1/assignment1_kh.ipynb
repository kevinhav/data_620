## Introduction

As a "beachhead" project, we decided to create a small network to represent the major U.S. cities we have been to. 

For this graph, our nodes are major U.S. cities, and the edges represent a abstracted version of highways that connect them

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Data Setup

Firstly, we will list out the major cities we've been to. These will be our **nodes**.We'll set this up as a list at first to make it easier to write, since we're doing this by hand. Our **edges** in this case are simplified geographic "connections", which could represent abstracted interstate highways.

Secondly, we will create a structure to index the cities turn them into proper node structures, including the `name`.

```python
# Write out list of cities
# Easiest to write out in east-to-west
cities = [
    "New York, NY",
    "Boston, MA",
    "Hartford, CT",
    "Albany, NY",
    "Buffalo, NY",
    "Columbus, OH",
    "Cleveland, OH",
    "Toledo, OH",
    "St. Joseph, MI",
    "Madison, WI",
    "New Orleans, LA",
    "Las Vegas, NV",
    "Los Angeles, CA",
]
```

```python
def create_nodes(cities: list):
    "Enrich a city list with colors to distinguish"

    nodes = []
    for i, city in enumerate(cities):
        node = (str(i), {"name": city})
        nodes.append(node)

    return nodes


# Create the nodes
nodes = create_nodes(cities=cities)
```

```python
# Create edges between cities (by index)
edges = [
    ("0", "3"),  # New York <-> Albany
    ("0", "4"),  # New York <-> Buffalo
    ("3", "4"),  # Albany <-> Buffalo
    ("5", "6"),  # Columbus <-> Cleveland
    ("5", "7"),  # Columbus <-> Toledo
    ("6", "7"),  # Cleveland <-> Toledo
    ("1", "2"),  # Boston <-> Hartford
    ("2", "0"),  # Hartford <-> New York
    ("4", "6"),  # Buffalo <-> Cleveland
    ("7", "8"),  # Toledo <-> St. Joseph
    ("8", "9"),  # St. Joseph<-> Madison
    ("10", "5"),  # New Orleans<-> Columbus
    ("11", "12"),  # Las Vegas <-> Los Angeles
    ("4", "10"),  # Buffalo <-> New Orleans
    ("4", "11"),  # Buffalo <-> Las Vegas
]
```

## Create the Network

We can now create the network using `networkx` library. To do this, we can use two simple methods; `add_nodes_from()` and `add_edges_from()` respectively.

We'll then draw the graph with `matplotlib` using city names as node labels.

```python
# Initialize the graph, then add nodes, edges
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)

# Specify the labels
labels = nx.get_node_attributes(G, "name")

# Draw the network
positions = nx.kamada_kawai_layout(G)
nx.draw(G, pos=positions, labels=labels, with_labels=True)

plt.show()
```
